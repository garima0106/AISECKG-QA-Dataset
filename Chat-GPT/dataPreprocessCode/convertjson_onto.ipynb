```python
import json
import ast
import os
```

```python
dir=os.getcwd()
dir
```

```python
filename = dir + '/datadump/Onto_extra'
```

```python
with open(filename, 'r') as fh:
    inputtxt= fh.read()
    #print(inputtxt)
```

```python
lines = inputtxt.split('\n')
res=[]
dic={}
method = "Ontology-based"
entity=None
relation=None
ontology=None
prompt=None
Q=None
A=None
qid=None
num=2142

for line in lines:
        line = line.strip()
        line=line.lstrip('0123456789.- ')
        line=line.replace('{', '')
        line=line.replace('}', '')
        line = line.strip()
        qid = 'A-'+str(num)
        if(line.startswith('"Prompt":')):
             prompt=line.replace('"Prompt":','')
             prompt = prompt.strip(" ")
        elif(line.startswith('"Entity":')):
             entity=line.replace('"Entity":','')   
             entity = entity.strip(" ")
             entity= entity.rstrip(',')
             entity = ast.literal_eval(entity)
        elif(line.startswith('"Relation":')):
              relation=line.replace('"Relation":','')
              relation = relation.strip(" ")
              relation= relation.rstrip(',')
              relation = ast.literal_eval(relation)
        elif(line.startswith('"Ontology":')):
              ontology=line.replace('"Ontology":','')
              ontology = ontology.strip(" ")
              ontology = ontology.rstrip(",")
              ontology = ast.literal_eval(ontology)
        elif(line.startswith('Q:') or line.startswith('"question:"')or line.startswith('"question"')):
              Q=line.replace('Q:','')
              Q= Q.strip()
              Q= Q.replace('"Question":  ',"")
              Q= Q.replace('"question":  ',"")
              Q= Q.replace('"question":',"")
              Q= Q.replace('"Question":',"")
              Q = Q.replace('"question" :',"")
              Q=Q.replace('"Question" :',"")
              Q=Q.replace('"question": ',"")
              Q=Q.replace('"Question": ',"")
              Q = Q.replace('"','')
              Q = Q.lstrip()
              Q= Q.rstrip(',')
        elif(line==""):
               continue
        
        elif(line.startswith('A:') or line.startswith('"answer:"')or line.startswith('"answer"')):    

            A=line.replace('A:','')
            A=A.replace('"Answer":',"")
            A=A.replace('"answer":',"")
            A=A.replace('"answer" :',"")
            A=A.replace('"Answer" :',"")
            A=A.replace('"answer": ',"")
            A=A.replace('"Answer": ',"")
            A = A.replace('"','')
            A = A.lstrip()
            dic["QID"]=qid
            dic["Method"]=method
            dic["Prompt"]=prompt
            dic["Entity"]=entity
            dic["Relation"]=relation
            dic["Ontology"]=ontology
            dic["Question"]=Q
            dic["Answer"]=A
            num=num+1
            #print(line)
            print(dic)
            #print("Next question and answer")
            print("\n") 
            res.append(dic)
            dic={} 
          #print(prompt,relation,entity)

# creating json file       
out_file = open("j2.json", "w")
json.dump(res, out_file, indent=4)
out_file.close()      


```
